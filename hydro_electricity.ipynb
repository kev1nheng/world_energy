```python
old_regime = pd.DataFrame(energy.loc[energy['iso_code'].isnull() == True])
```

```python
energy = pd.DataFrame(energy.loc[energy['iso_code'].isnull() != True])
```

```python
energy.isnull().sum()/len(energy)*100
```

```python
energy.fillna(0,inplace= True)
```

```python
energy.info()
```

```python
energy.isnull().sum()/len(energy)*100
```

```python
energy.dtypes
```

```python
canada_energy = pd.DataFrame(energy.loc[energy['iso_code'] == 'CAN'])
```

```python
canada_energy.drop(columns=['iso_code','country'],inplace = True)
```

```python
canada_energy.set_index('year',inplace = True)
```

```python
canada_energy.dtypes
```

```python
for i in ('coal_prod_change_pct','coal_prod_change_twh'):
    canada_energy[i].plot(legend=('coal_prod_change_pct','coal_prod_change_twh'))
```

```python
for i in ('hydro_share_elec','hydro_cons_change_pct','hydro_share_energy','hydro_cons_change_twh','hydro_consumption','hydro_elec_per_capita','hydro_energy_per_capita'):
    canada_energy[i].plot(legend=('hydro_share_elec',
                                  'hydro_cons_change_pct',
                                  'hydro_share_energy',
                                  'hydro_cons_change_twh',
                                  'hydro_consumption','hydro_elec_per_capita','hydro_energy_per_capita'))
```
